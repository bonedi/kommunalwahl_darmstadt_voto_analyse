```python
import pandas as pd
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram
from matplotlib.colors import ListedColormap
```

```python
# Die Daten sind im Longformat gespeichert
# Wir brauchen sie aber im Breitformat (jede Zeile eine Partei, jede Spalte eine Frage)
df_long = pd.read_excel("data/votes.xlsx")
df = df_long.pivot(columns='shorttitle', index='parteien', values='votes')
```

```python
own_cmap = ListedColormap(["#d7191c",'#fdae61',  '#ffffbf', '#abd9e9','#2c7bb6'])
```

```python
cg = sns.clustermap(df, cmap = own_cmap, col_cluster=True,)
col_order = cg.data2d.columns
ax = cg.ax_heatmap

# Setze alle x-Tick-Labels
cg.ax_heatmap.set_xticks(range(len(col_order)))
cg.ax_heatmap.set_xticklabels(col_order, rotation=90)
```

```python
cg.savefig("plots/clustermap_parteien_x_fragen.png")
```
